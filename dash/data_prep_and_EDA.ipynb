Data: https://data.london.gov.uk/dataset/number-bicycle-hires

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 8)
```

# Data Preparation

```python
df = pd.read_excel('tfl-daily-cycle-hires.xls', sheet_name='Data', usecols=['Day', 'Number of Bicycle Hires'], index_col=0)
df.index.rename('DateTime', inplace=True)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.to_csv('cycle-hires-prepared.csv')
```

# Exploratory Data Analysis

Graphs to recreate in dashboard.

```python
df.plot();
```

```python
df.hist(bins=20);
```

```python
ax = df.groupby(df.index.month).mean().plot(kind='bar')
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax.set_xlabel('Month');
```

```python
ax = df.groupby(df.index.weekday).mean().plot(kind='bar')
ax.set_xticklabels(['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])
ax.set_xlabel('Day of week');
```
